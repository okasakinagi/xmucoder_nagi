### 1. 列表操作
**题目**：创建一个包含10个随机整数的列表（范围1-100），编写代码实现：
- 计算列表所有元素的和
- 找出最大值和最小值
- 将列表逆序排列

```python
import random

#your code
array = []
sum =0
min,max = 101,-1
for i in range(10):
    temp = random.randint(1,100)
    array.append(temp)
    sum += temp
    max = max if temp<max else temp
    min = min if temp>min else temp


print(f"原始列表：{array}")
print(f"总和：{sum}")
print(f"最大值：{max} 最小值：{min}")

array = array[::-1]
print(f"逆序列表{array}")
```

### 2. 字典应用
**题目**：创建一个字典，键为学生姓名（至少5个），值为对应的分数（0-100）。编写代码：
- 计算平均分
- 找出最高分和最低分的学生
- 添加一个新学生并更新字典

```python
#your code
dictionary = {"小明":95,"小红":96,"小白":97,"小美":91,"小新":96,"小小":94}

sum = 0
average = 0
max_stu_name,max_stu_grade = None,-1 
min_stu_name,min_stu_grade = None,101
add_stu_name,min_stu_grade = "小黑",99

for name,grade in dictionary.items():
    sum += grade
    if(grade>max_stu_grade):
        max_stu_grade = grade
        max_stu_name = name
    if(grade<min_stu_grade):
        min_stu_grade = grade
        min_stu_name = name

average = sum/len(dictionary)
print(f"学生总数：{len(dictionary)} 平均分：{average}")
print(f"最高分：{max_stu_name} {max_stu_grade} 最低分：{min_stu_name} {min_stu_grade}")

print(f"原学生成绩字典：{dictionary}")
dictionary[add_stu_name] = min_stu_grade
print(f"添加学生后学生成绩字典：{dictionary}")
```

### 4. 条件与循环（素数）
**题目**：打印100以内的所有素数（质数）。

```python
#your code
import math

def prime(n:int)->bool:
    for i in range(2,math.ceil(math.pow(n,0.5))+1):
        if(n%i==0):
            return False
    return True

print("100以内素数有：")
for i in range(2,101):
    if prime(i):
        print(i,end=' ')
```

### 5. 回文判断函数
**题目**：编写一个函数 `is_palindrome(s)`，判断字符串 `s` 是否是回文（正反读相同），忽略大小写和非字母数字字符。


```python
#your code
def is_palindrome(s:str)->bool:
    s = s.lower()
    return s == s[::-1]

#test
test_strs = ["A man a plan a canal Panama", "Hello", "racecar", "Python", "Level"]
for test_str in test_strs:
    if is_palindrome(test_str):
        print(f'"{test_str}" 是回文字符串')
    else:
        print(f'"{test_str}" 不是回文字符串')
```

### 7. 可变参数函数
**题目**：编写函数 `multiply_all(*args)`，接受任意个参数并返回它们的乘积。


```python

#your code
def multiply_all(*args):
    result = 1
    for num in args:
        result *= num
    return result

# 测试
print(multiply_all(2, 3, 4))    # 24
print(multiply_all(5, 6))       # 30
print(multiply_all(10))         # 10
```

### 8. 商和余数
**题目**：定义一个函数`divide(a, b)`，返回整数`a`除整数`b`的商和余数。如果输入非法（除零和类型错误）则捕获 `ValueError` 异常并提示重新输入。


```python
#your code
def divide(a,b):
    try:
        return a/b,a%b
    except:
        return False,None
    
#test
while(1):
    a,b = int(input("请输入被除数: ")),int(input("请输入除数: "))
    print(f"当前要求解的除法表达式是 {a} / {b}")
    result,mod = divide(a,b)
    if(result==False):
        print("除数不能为0，请重新输入")
    else:
        print(f"{a} 除以 {b} 的商是 {result} 余数是 {mod}")
        break
```

## 四、文件读写

### 9. 逐行读取统计
**题目**：读取一个`utf-8`文本文件（`data`目录中的文件 `books.txt`），统计文件包含的行数和字符数。


```python
#your code
with open("data/books.txt","r",encoding="UTF-8") as file:
    line_count =0
    char_count =0
    for line in file.readlines():
        line_count +=1
        char_count += len(line)
    print(f"文件共有 {line_count} 行, {char_count} 个字符")
```
