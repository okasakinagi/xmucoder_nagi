### 题目1：数组分块求和
随机创建一个形状为 (4, 4) 的整数数组（元素范围0-100）。将其划分为 2x2 的子数组，并计算每个子数组的和。



```python
# 设置随机种子以便结果可重现
np.random.seed(42)
array = np.random.randint(0, 101, size=(4, 4))
print(array)
array_splitted = map(np.array_split,np.array_split(array, 2),[2,2],[1,1])
array_splitted = array_splitted.__next__()+array_splitted.__next__()
print(np.array(list(map(np.sum,array_splitted))).reshape(2,2))
```

### 题目2：数组滑动窗口
随机生成一个长度为10的一维整数数组（元素范围0-100）。计算大小为 3 的滑动窗口的平均值（取整）。


```python
# 设置随机种子以便结果可重现
np.random.seed(42)
array = np.random.randint(0, 101, size=(1,10))
print(f"原始数组: {array}")
array_mean = np.array([int(np.mean(array[:,i:i+3])) for i in range(array.shape[1]-2)])
print(f"滑动窗口平均值（整数）: {array_mean}")

```

### 题目3：数组对角线填充
给定一个形状为 (5, 5) 的数组，将其主对角线元素填充为 1，对角线两侧的元素填充为 2，其余元素为 0。

```python

array = np.eye(5).flatten()
for i in range(array.shape[0]):
    if array[i] == 1:
        if i > 0:
            array[i-1] = 2
        if i < array.shape[0] - 1:
            array[i+1] = 2
print(array.reshape(5, 5))
```

### 题目4：数组条件替换

### 题目5：数组分组统计
随机生成一个长度为9的整数数组（元素范围0-9），随机生成一个长度为9的分组索引，索引取值为0，1，2。求每组的平均值

### 题目6：数组元素筛选

### 题目7：数组统计计算
创建一个4×5的随机整数数组（元素范围10-50），计算：
- 所有元素的平均值、中位数
- 每列的最大值、每行的最小值
- 数组的标准差、方差
- 数组元素总和

```python
# 设置随机种子以便结果可重现
np.random.seed(42)
array = np.random.randint(10,51,size=(4,5))
print(f"随机数组: {array}")
print("平均值: %.2f" % np.mean(array))
print("中位数: %.2f" % np.median(array))
print(f"每列最大值: {np.max(array, axis=0)}")
print(f"每列最小值: {np.min(array, axis=0)}")
print("标准差: %.2f" % np.std(array))
print("方差: %.2f" % np.var(array))
print(f"元素总和: {np.sum(array)}")
```

### 题目8：数组归一化
随机生成一个长度为10的整数数组（元素范围0-9），将其归一化到 [0, 1] 范围，小数点后保留2位。


```python
# 设置随机种子保证结果可重现
np.random.seed(42)
array = np.random.randint(0,10,size=(1,10))
print(f"原始随机数组: {array}")
array_min = np.min(array)
array_max = np.max(array)
array = np.array(list(map(lambda x: (x-array_min)/(array_max-array_min), array.flatten()))).round(decimals=2)
print(f"归一化后的数组: {array}")
```

### 题目9：数组逻辑运算
给定两个布尔数组 a = [True, False, True] 和 b = [False, False, True]，计算它们的逻辑与、逻辑或和数组a的逻辑非。


```python
a = np.array([True, False, True])
b = np.array([False, False, True])
print(f"逻辑与: {np.logical_and(a, b)}")
print(f"逻辑或: {np.logical_or(a, b)}")
print(f"逻辑非: {np.logical_not(a)}")
```

### 题目10：数组拼接与拆分
给定两个数组 a = [1, 2, 3, 4] 和 b = [5, 6, 7, 8]，将它们按行拼接成一个新数组，然后将新数组再水平拆分（按列拆分）为两个子数组。


```python

a = np.array([[1, 2, 3 ,4]])
b = np.array([[5, 6, 7, 8]])
a_b = np.concatenate((a,b), axis=0)
print(f"拼接后数组: \n{a_b}")
print(f"拆分后数组1：\n{a_b[:,:2]}")
print(f"拆分后数组2：\n{a_b[:,2:]}")
```

### 题目11：数组标准化
随机生成一个长度为10的整数数组（元素范围10-50），将其z-score标准化，计算标准化后的数据的均值和标准差，小数点后保留2位。。


```python

# 设置随机种子保证结果可重现
np.random.seed(42)
array = np.random.randint(10, 51, size=(1, 10))
print(f"原始数组: {array}")
array_zscore = np.array(list(map(lambda x: (x-np.mean(array))/np.std(array), array.flatten()))).round(decimals=2)
print(f"标准化后的数组: {array_zscore}")
print("标准化后均值: %.2f" % np.mean(array_zscore))
print("标准化后标准差: %.2f" % np.std(array_zscore))

```
